CÂU 13

```python
df['Normalized_age'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
print(df)
```

CÂU 14

```python
def phanloai_age(age):
    if age <= 30:
        return 'Young'
    elif age <= 60 and age > 30:
        return 'Middle-aged'
    else:
        return 'Old'
```

```python
df['Age_group'] = df['Age'].apply(phanloai_age)
```

```python
print(df)
```
